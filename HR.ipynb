```python
import matplotlib.pyplot as plt
import pandas as pd 
from sqlalchemy import create_engine
from config import dbpass
engine = create_engine(f'postgresql://postgres:{dbpass}@localhost:5432/HR')
db = engine.connect()
```

```python
cursor = db.execute("SELECT title, salary FROM vwEmployeeData")  
tupples = cursor.fetchall()
cursor.close()
df_empdata = pd.DataFrame(tupples, columns=['title','salary'])
df_empdata.head()
```

```python
df=df_empdata
plt.hist(df['salary'], bins=60)
plt.show()
```

```python
df=df_empdata
df=df.groupby('title').mean()
df
```

```python
plt.figure(figsize=(24,12))
plt.bar( df.index , df['salary'])
plt.show()
```
